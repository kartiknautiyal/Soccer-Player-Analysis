#### Looking at the conjectures, it was clear that some features were related to the target variable. Age, number of minutes played and the position of the player were found to be the most important predictors. Multiple models were trained by using these as the input variables and goals as the target variables. Since position of a player was a categorical variable, it could not be directly included. Considering that forwards end up scoring more goals than midfielders, defenders and goalkeepers, it was considered as an ordinal variable and each position was assigned a number based on this. Model training was done using a training set which was 80% of the data, and 20% of the data was kept as the validation set.


```python
# write here

# Popularity score calculation 

score = pd.DataFrame()
score["Name"] = dff["Name"]
score["Position"] = dff["Position"]
score["mins"] = (dff["Mins"]- dff["Mins"].min())/(dff["Mins"].max()- dff["Mins"].min())
score["perf"] = (dff["perf"]- dff["perf"].min())/(dff["perf"].max()- dff["perf"].min())
score["goals"] = (dff["Goals"]- dff["Goals"].min())/(dff["Goals"].max()- dff["Goals"].min())
#score["assists"] = (dff["Assists"]- dff["Assists"].min())/(dff["Assists"].max()- dff["Assists"].min())
score["yellow_cards"] = (dff["Yellow_Cards"]- dff["Yellow_Cards"].min())/(dff["Yellow_Cards"].max()- dff["Yellow_Cards"].min())
score["red_cards"] = (dff["Red_Cards"]- dff["Red_Cards"].min())/(dff["Red_Cards"].max()- dff["Red_Cards"].min())
score["stad_cap"] = (dff["stad_cap"]- dff["stad_cap"].min())/(dff["stad_cap"].max()- dff["stad_cap"].min())

score["total"] = score["mins"]*5 + score["perf"]*5 + score["goals"]*5 - score["yellow_cards"]*3 -score["red_cards"]*5 +score["stad_cap"]*10
score["Popularity_score"]= ((score["total"]- score["total"].min())/(score["total"].max()- score["total"].min())*100)


def score_calculator(mins, perf, goals, yellow, red, stad_cap):
    mins_score = (mins- dff["Mins"].min())/(dff["Mins"].max()- dff["Mins"].min())
    perf_score = (perf- dff["perf"].min())/(dff["perf"].max()- dff["perf"].min())
    goal_score = (goals- dff["Goals"].min())/(dff["Goals"].max()- dff["Goals"].min())
    yc_score = (yellow- dff["Yellow_Cards"].min())/(dff["Yellow_Cards"].max()- dff["Yellow_Cards"].min())
    rc_score = (red- dff["Red_Cards"].min())/(dff["Red_Cards"].max()- dff["Red_Cards"].min())
    stad_score = (stad_cap- dff["stad_cap"].min())/(dff["stad_cap"].max()- dff["stad_cap"].min())
    total = mins_score*5 + perf_score*5 + goal_score*5 - yc_score*3 - rc_score*5 + stad_score*10
    final = ((total - score["total"].min())/(score["total"].max()- score["total"].min())*100)
    return(final)
```

```python
df12 = score.nlargest(10, columns=['Popularity_score'])
df12 = df12[['Name', 'Popularity_score']]
plt.figure(figsize=(18,8))
plt.bar(x=df12['Name'], height=df12['Popularity_score'], color=['midnightblue', 'navy', 'darkblue', 'mediumblue', 'blue', 'royalblue', 'cornflowerblue', 'dodgerblue', 'deepskyblue', 'lightskyblue'])
plt.title("Top 10 Popular Players", fontsize=12)
plt.xlabel("Name", fontsize=12)
plt.ylabel("Popularity Score", fontsize=12)
plt.show()
```
